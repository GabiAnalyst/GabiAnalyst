## ANSWERING THE QUESTIONS GIVEN.

```python
# extract year and year-month from date column.
df['Year'] = pd.to_datetime(df['Date Added']).dt.year
```

```python
# To calculate the number of days from release to today.
df['Year-Month']=pd.to_datetime(df['Date Added']).dt.to_period('M')
```

```python
df.head()
```

```python
# to calculate the number of days from release till today.
df['Days from release']= (pd.Timestamp.today()-df['Date Added'])
```

```python
df.head()
```

```python
# a function to Tv-Series duration
def extract_tv_duration(Duration):
        if 'Season' in Duration:
            return Duration.split(' ')[0]
        else:
            return np.nan
```

```python
#creating new column for Tv-Series Duration.
df['Tv Duration'] = df['Duration'].apply(extract_tv_duration)
```

```python
df.head()
```

```python
# to generate random ratings
def generate_random_rating():
    return round(np.random.uniform(0, 10), 2)
```

```python
# creating column for random rating.
df['Random Rating']=df.apply(lambda x: generate_random_rating(), axis=1)
```

```python
df.head()
```

```python
df['Genres']= df['Listed_In'].apply(lambda x: x.split(', '))
```

```python
# using counter to find the most common genres.
Genre_counter = Counter([genre for sublist in df['Genres'] for genre in sublist])
Most_common_genre = Genre_counter.most_common(1)[0][0]
```

## DATA VISUALIZATION

```python
plt.figure(figsize=(10,6))
sns.barplot(x=Count_Director, y= Count_Director.index, palette='viridis')
plt.xlabel('Number of Movies')
plt.ylabel('Director')
plt.title('Top 10 most Popular Directors')
plt.show()
```

```python
Count_Country = df['Country'].value_counts()
Top_Countries = Count_Country.head(5)
Bottom_Countries = Count_Country.tail(5)
```

```python
# Top 5 Countries.
plt.figure(figsize=(10,6))
sns.barplot(x=Top_Countries, y=Top_Countries.index, palette='muted')
plt.xlabel('Number of Movies')
plt.ylabel('Country')
plt.title('Top 5 Countries Producing Movies')
plt.show()
```

```python
# bottom five Countries.
plt.figure(figsize=(10,6))
sns.barplot(x=Bottom_Countries, y=Bottom_Countries.index, palette='muted')
plt.xlabel('Number of Movies')
plt.ylabel('Country')
plt.title('Bottom 5 Countries Producing Movies')
plt.show()
```

```python
# Movie trends over the years.
year_counts = df['Year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o')
plt.xlabel('Year')
plt.ylabel('Number of Movies Added')
plt.title('Movie Trends Over the Years')
plt.show()
```

```python
#Distribution of Movie Duration.
df['duration_numeric'] = df['Duration'].str.extract('(\d+)')
df['duration_numeric'] = pd.to_numeric(df['duration_numeric'], errors='coerce')

# Plotting distribution of movie durations
plt.figure(figsize=(10, 6))
sns.histplot(df['duration_numeric'].dropna(), bins=30, kde=True, color='skyblue')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.title('Distribution of Movie Durations')
plt.show()
```

```python
# Average movie duration by country
avg_duration_by_country = df.groupby('Country')['duration_numeric'].mean().sort_values(ascending=False).dropna().head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_duration_by_country.values, y=avg_duration_by_country.index, palette='rocket')
plt.xlabel('Average Duration (minutes)')
plt.ylabel('Country')
plt.title('Top 10 Countries with Highest Average Movie Duration')
plt.show()
```

```python
# Most common Genre
all_genres = [genre for sublist in df['Genres'] for genre in sublist]
genre_counter = Counter(all_genres)
top_genres = genre_counter.most_common(10)


plt.figure(figsize=(10, 6))
sns.barplot(x=[genre[1] for genre in top_genres], y=[genre[0] for genre in top_genres], palette='mako')
plt.xlabel('Number of Movies')
plt.ylabel('Genre')
plt.title('Top 10 Most Common Genres')
plt.show()
```
